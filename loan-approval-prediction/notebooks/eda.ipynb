# Loan Approval Prediction — Exploratory Data Analysis

Quick look at the dataset before modeling: shape, missing values, distributions, and how each feature relates to `Loan_Status`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline

df = pd.read_csv("../data/loan_data.csv")
df.shape
```

## 1. First look at the data

```python
df.head()
```

```python
df.info()
```

## 2. Missing values

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

## 3. Target distribution

```python
df['Loan_Status'].value_counts().plot(kind='bar', color=['#4C72B0', '#DD8452'])
plt.title('Loan Status Distribution (Y=Approved, N=Rejected)')
plt.xticks(rotation=0)
plt.show()

df['Loan_Status'].value_counts(normalize=True)
```

## 4. Numeric feature distributions

```python
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
fig, axes = plt.subplots(2, 2, figsize=(11, 8))
for ax, col in zip(axes.ravel(), numeric_cols):
    df[col].hist(bins=30, ax=ax, color='#4C72B0')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 5. Income vs approval status

```python
sns.boxplot(data=df, x='Loan_Status', y='ApplicantIncome')
plt.title('Applicant Income by Loan Status')
plt.show()
```

## 6. Approval rate by credit history

```python
pd.crosstab(df['Credit_History'], df['Loan_Status'], normalize='index').plot(
    kind='bar', stacked=True, color=['#DD8452', '#4C72B0'])
plt.title('Approval Rate by Credit History')
plt.xticks(rotation=0)
plt.legend(['Rejected', 'Approved'])
plt.show()
```

## 7. Approval rate by property area

```python
pd.crosstab(df['Property_Area'], df['Loan_Status'], normalize='index').plot(
    kind='bar', stacked=True, color=['#DD8452', '#4C72B0'])
plt.title('Approval Rate by Property Area')
plt.xticks(rotation=0)
plt.legend(['Rejected', 'Approved'])
plt.show()
```

## 8. Categorical feature breakdown

```python
cat_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, col in zip(axes.ravel(), cat_cols):
    pd.crosstab(df[col], df['Loan_Status'], normalize='index').plot(
        kind='bar', stacked=True, ax=ax, color=['#DD8452', '#4C72B0'], legend=False)
    ax.set_title(col)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
axes.ravel()[-1].axis('off')
fig.legend(['Rejected', 'Approved'], loc='lower right')
plt.tight_layout()
plt.show()
```

## 9. Correlation among numeric features

```python
corr = df[numeric_cols + ['Credit_History']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

## Takeaways

- `Credit_History` shows the clearest split between approved and rejected applicants.
- Higher combined income and `Education = Graduate` are mildly associated with approval.
- Several columns have missing values that need imputing before modeling (`Credit_History`, `Self_Employed`, `LoanAmount`, etc.) — handled in `src/model.py`.
